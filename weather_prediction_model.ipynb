# Creating Model

```python
df1["TARGET"] = df1.shift(-1)["TMAX"]   #predicting tomorrows data from today
```

```python
df1
```

```python
df1 = df1.iloc[:-1,:].copy()
```

```python
df1
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)
```

```python
predictors = ['PRCP','TMAX','TMIN']
```

```python
train = df1.loc[:"2023-05-15"]
test = df1.loc["2023-05-16":]
```

```python
test
```

```python
reg.fit(train[predictors], train["TARGET"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(test["TARGET"], predictions)
```

```python
combined = pd.concat([test["TARGET"], pd.Series(predictions, index=test.index)], axis=1)
combined.columns = ["ACTUAL", "PREDICTED"]
```

```python
combined
```

```python
combined.plot()
```

```python
reg.coef_
```

```python
def make_predictions(predictors, df1, reg):
    train = df1.loc[:"2023-05-15"]
    test = df1.loc["2023-05-16":]
    reg.fit(train[predictors], train["TARGET"])
    predictions = reg.predict(test[predictors])
    error = mean_squared_error(test["TARGET"], predictions)
    combined = pd.concat([test["TARGET"], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ["ACTUAL", "PREDICTED"]
    return error, combined
```

Adding more predictors

```python
df1["MONTH_MAX"] = df1["TMAX"].rolling(30).mean()

# df1["month_day_max"] = df1["month_max"] / df1["temp_max"]

# df1["max_min"] = df1["temp_max"] / df1["temp_min"]
```

We need before 30 days to predict MONTH_MAX therefore getting NaN values on starting dates

```python
df1["MONTH_DAY_MAX"] = df1["MONTH_MAX"] / df1["TMAX"]

df1["MAX_MIN RATIO"] = df1["TMAX"] / df1["TMIN"]
```

```python
df1
```

```python
predictors = ["PRCP", "TMAX", "TMIN", "MONTH_DAY_MAX", "MAX_MIN RATIO"]

error, combined = make_predictions(predictors, df1, reg)
error
```
